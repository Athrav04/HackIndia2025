```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Combine relevant text fields for similarity matching
df['product_text'] = df['name'] + " " + df['description']

# Create TF-IDF vectors for text similarity
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['product_text'])
```

```python
def find_similar_products(input_product, df, tfidf_matrix, vectorizer, top_n=5):
    """
    Find top_n most similar eco-friendly products to the input product

    Args:
    input_product (dict): Product details from the website with keys like 'name', 'description', etc.
    df: Your greenDB DataFrame
    tfidf_matrix: Precomputed TF-IDF matrix
    vectorizer: Fitted TF-IDF vectorizer
    top_n: Number of similar products to return

    Returns:
    DataFrame with top_n similar products
    """
    # Combine input product text in same way as training data
    input_text = input_product.get('name', '') + " " + \
                 input_product.get('description', '')

    # Vectorize the input text
    input_vector = vectorizer.transform([input_text])

    # Calculate cosine similarities
    similarities = cosine_similarity(input_vector, tfidf_matrix)

    # Get indices of top_n most similar products
    similar_indices = similarities.argsort()[0][-top_n:][::-1]

    # Return the most similar products
    return df.iloc[similar_indices].copy()
```

```python
def rank_products(similar_products):
    """
    Rank similar products by multiple criteria to find the "best" one

    Args:
    similar_products: DataFrame of similar products

    Returns:
    The single best product to highlight
    """
    # Add scoring for different factors
    similar_products['score'] = 0

    # Higher price might mean better quality (adjust weights as needed)
    similar_products['score'] += similar_products['price'] * 0.1

    # More sustainability labels is better
    similar_products['score'] += similar_products['sustainability_labels'].str.count(',') * 10

    # Higher in search results is better
    similar_products['score'] += (len(similar_products) - similar_products.index) * 5

    # Return the product with highest score
    return similar_products.loc[similar_products['score'].idxmax()]
```

```python
example_product = {
    "name": "Classic Straight Fit Jeans for Men",
    "description": "Mid-waist straight fit jeans made with 100% cotton. Five-pocket styling with zip fly and button closure. Machine washable.",
    "brand": "FashionHub",
    "category": "Clothing > Men > Jeans",
    "price": 45.99,
    "currency": "USD"
}
```

```python
similar_eco_products = find_similar_products(example_product, df, tfidf_matrix, vectorizer)
similar_eco_products
```
